```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-2*np.pi,2*np.pi,256)
c,s = np.cos(x),np.sin(x)
```

```python
plt.plot(x,c)
plt.plot(x,s)
# plt.plot(s,c)
```

# 对轴线、刻度、刻度范围的操作
xaxis
xlim
xtick

```python

fig,a = plt.subplots(figsize = (12,8))
a.plot(x,c,color='red',lw=2,ls='--')
a.plot(x,s,color='green',lw=2,ls=':')
a.set_xlim(-10,10)
a.set_ylim(-2.0,2.0)
a.set_xticks(np.linspace(-8,8,9)) ;

```

```python
fig,b = plt.subplots(figsize = (10,6))
b.plot(x,c,color='red',lw=2,ls='--')
b.plot(x,s,color='green',lw=2,ls=':')
b.set_xlim(x.min()*1.1,x.max()*1.1)  
b.set_ylim(c.min()*1.1,c.max()*1.1) 
b.set_xticks(np.linspace(-6,6,9));
```

```python
fig,b = plt.subplots(figsize = (10,6))
b.plot(x,c,color='red',lw=2,ls='--')
b.plot(x,s,color='green',lw=2,ls=':')
b.set_xticks(np.linspace(-6,6,9));

xmin ,xmax = x.min(), x.max()
ymin, ymax = c.min(), c.max()

dx = (xmax - xmin) * 0.2
dy = (ymax - ymin) * 0.2

b.set_xlim(xmin - dx, xmax + dx)
b.set_ylim(ymin - dy, ymax + dy);
```

```python
fig,b = plt.subplots(figsize = (10,6))
b.plot(x,c,color='red',lw=2,ls='--')
b.plot(x,s,color='green',lw=2,ls=':')
b.set_xticks((-np.pi, -np.pi/2, 0, np.pi/2, np.pi) )
b.set_yticks((-1, 0, +1));
```

```python
fig,c = plt.subplots(figsize = (10,6))
c.plot(x,c,color='red',lw=2,ls='--')
c.plot(x,s,color='green',lw=2,ls=':')
c.set_xticks((-np.pi, -np.pi/2, 0, np.pi/2, np.pi),(r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$')
c.set_yticks((-1, 0, +1),(r'$-1$', r'$0$', r'$+1$')
```
